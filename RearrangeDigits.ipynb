```python
def rearrangeDigitsThreeDigits(listDigits):
    listAllVersions = []
    
    for i in range(3):
        newArray = listDigits
        listAllVersions.append(list(newArray))
        
        # First rearrange last two
        
        temp = newArray[2]
        newArray[2] = newArray[1]
        newArray[1] = temp
        #print(listAllVersions)
        listAllVersions.append(list(newArray))
        #print(listAllVersions)
        newArray[1] = newArray[0]
        newArray[0] = temp
        
    return listAllVersions
rearrangeDigitsThreeDigits([1, 2, 3])
```

```python
def arraysPlaceFirstDig(arrOfArr, firstDig):
    finalResults = []
    for i in arrOfArr:
        for j in range(0, len(i) + 1):
            temp = i
            temp.insert(j, firstDig)
            finalResults.append(list(temp))
            del temp[j]
    return finalResults
arraysPlaceFirstDig([[2, 3, 6], [3, 2, 6]], 5)
```

```python
def rearrangeDigits(digits):
    if len(digits) == 3:
        arr = rearrangeDigitsThreeDigits(digits)
        return arr
    firstDigit = digits[0]
    del digits[0]
    resultsNMO = rearrangeDigits(digits)
    return arraysPlaceFirstDig(resultsNMO, firstDigit)
rearrangeDigits([1, 2, 3, 4, 5 ,6])
```
