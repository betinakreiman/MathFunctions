```python
def gcd(a, b):
    if a%b == 0:
        return b
    return gcdWithRecursion(b, a%b)
gcd(49, 35)
```

```python
def howManyBelowAreRelativelyPrime(n):
    sumRP = 1
    for i in range(2, n):
        if gcd(i, n) == 1:
            sumRP = sumRP + 1
    return sumRP
howManyBelowAreRelativelyPrime(20)
```

```python
import matplotlib.pyplot as plt

def graphKylesThing(n):
    nArr = []
    relativePArr = []
    for i in range(1, n):
        nArr.append(i)
        relativePArr.append(howManyBelowAreRelativelyPrime(i))
        
    plt.plot(nArr, relativePArr)
    plt.title('Kyles Thing')
    plt.xlabel('n')
    plt.ylabel('number of number relatively prime')
    plt.show()
graphKylesThing(10000)
```
