```python
def makingTonsOfNumbersForFunc(maxK, maxZ):
    listOfAnswers = []
    for i in range(1, maxK):
        for j in range(i+1, maxZ):
            answer = 3 * i * pow(j, 2)
            answer = answer - 3 * pow(i, 2) * j
            answer = answer + pow(i, 3)
            tempList = [answer, i, j]
            
            listOfAnswers.append(tempList)
    return listOfAnswers
newList = makingTonsOfNumbersForFunc(100, 1000)
print(newList)
```
